```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
text_documents = text_splitter.split_documents(text_documents)
text_documents[:10]
```

```python
import os
import deepai

# Set your DeepAI API key
deepai.api_key = os.getenv("DEEP_AI_API_KEY")  # Ensure the environment variable is set

# Example texts to embed
texts = text_documents[:30]

# Function to get embeddings from DeepAI
def get_deepai_embeddings(texts):
    embeddings = []
    for text in texts:
        response = deepai.call_standard_api('text-embeddings', {
            'text': text
        })
        embeddings.append(response['output'])  # Adjust based on the actual response structure
    return embeddings

# Get embeddings
embeddings = get_deepai_embeddings(texts)
```
